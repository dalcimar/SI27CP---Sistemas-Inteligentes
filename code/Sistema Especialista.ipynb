```python
from sklearn import datasets
breast_cancer = datasets.load_breast_cancer()

X = breast_cancer.data[:, :2]  # we only take the first two features.
y = breast_cancer.target
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=123, stratify=y)
```

```python
class Especialist():
    def __init__(self, feature, rule):
        self.feature = feature
        self.rule = rule

    def fit(self, X, y):
        self.feature = 0
        self.rule = 10
        
    def predict(self, X):  
        y_hat = np.where(X[:,self.feature]>self.rule, 0, 1)
        return y_hat
        
    def score(self, y, y_hat):
        accuracy = np.sum(y==y_hat) / np.size(y)
        return accuracy
```

```python
model = Especialist(0,15)
```

```python
# train analise
model.fit(X_train, y_train)

y_pred = model.predict(X_train)
model.score(y_train,y_pred)
```
