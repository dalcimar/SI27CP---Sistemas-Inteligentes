```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

import scipy as sp
```

```
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```
import seaborn as sns
cov = np.cov(X.T)
sns.heatmap(cov, cmap='viridis')
cov
```

```
import seaborn as sns
corr = np.corrcoef(X.T)
sns.heatmap(corr, vmin=-1, vmax=1, cmap='viridis')
corr
```
